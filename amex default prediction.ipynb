```python
import os
#import pandas as pd
import cudf as pd
from pycaret.regression import *
```

```python
root="/amex/"
headers=pd.read_csv(root+"headers.csv")
headers = headers.columns
labels =pd.read_csv(root+"train_labels.csv")
```

```python
#!split -l 75000 train_data.csv
```

```python
trains = []
for r, d, f in os.walk(root):
    for file in f:
#        if "traina" in file or "trainb" in file:
        if "xa" in file or "xb" in file or "xc" in file:
            trains.append(os.path.join(root, file))
```

```python
def buildModel(train, model):
    s = setup(data=train, target='target', fold_strategy='timeseries', fold=3, use_gpu=True, silent = True, session_id = 123,
             #imputation_type='iterative', #ignore_low_variance=True, combine_rare_levels=True, 
              profile=True)
    best = compare_models(include = ['ridge']) #, 'par', 'lightgbm', 'catboost', 'gbr'])
#    evaluate_model(best)
#    predict_model(best)
    save_model(best, "./models"+model)
```

```python
traindfs = []
#for i in range(len(trains)):
for i in range(2):
    print(trains[i])
#    traindfs.append(pd.read_csv(trains[i], names=headers, header=None))
#    traindfs[i] = traindfs[i].merge(labels, on="customer_ID", how="inner")
    traindf = pd.read_csv(trains[i], names=headers, header=None)
    traindf = traindf.merge(labels, on="customer_ID", how="inner")
    buildModel(traindf, trains[i])
```
